```python

data={ "month":['januray','februaru','march','april','may'],
       "sales":[10000,12000,13000,14000,16000],
        "profit":[1000,2000,3000,4000,5000]
     }

df=pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt
```

# LINE PLOT


```python
plt.figure(figsize=(6,3))
plt.xlabel('month')
plt.ylabel('sales($)')
plt.title('sales over month')
plt.plot(df['month'],df['sales'],label='sales',marker='o',linestyle='-',color='blue')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.xlabel('month')
plt.ylabel('sales($)')
plt.title('sales over month')
plt.plot(df['month'],df['sales'],label='sales',marker='o',linestyle=':',color='blue')
plt.legend()
plt.show()

# label means what to show in a graph
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y1 = [0, 1, 4, 9]
y2 = [0, 1, 2, 3]
y3 = [9, 4, 1, 0]

plt.plot(x, y1, linestyle='-', label='Solid')
plt.plot(x, y2, linestyle='--', label='Dashed')
plt.plot(x, y3, linestyle=':', label='Dotted')

plt.legend()
plt.title("Line Style Examples")
plt.show()
```

# BAR PLOT MONTH VS PROFIT__


```python
plt.figure(figsize=(7,5))
width=0.3
plt.xlabel('month')
plt.ylabel('sales($)')
plt.title('monthly sales & profits')
plt.bar(df['month'],df['sales'],width=width,linestyle='-',color='skyblue',label='sales')
plt.bar(df['month'],df['profit'],width=width,linestyle='--',color='green',label='profit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt


data={ "month":['januray','februaru','march','april','may'],
       "sales":[10000,12000,13000,14000,16000],
        "profit":[1000,2000,3000,4000,5000]
     }

df=pd.DataFrame(data)

def generate_plot(plot_type):
  fig=plt.figure(figsize=(8,5))

  if plot_type == 'line plot':
    plt.xlabel('month')
    plt.ylabel('sales($)')
    plt.title('sales over month')
    plt.plot(df['month'],df['sales'],label='sales',marker='o',linestyle='-',color='blue')
    plt.legend()



  elif plot_type == 'stacked bar chart':
    fig.set_size_inches(10, 6)
    width=0.3
    plt.xlabel('month')
    plt.ylabel('sales($)')
    plt.title('monthly sales & profits')
    plt.bar(df['month'],df['sales'],width=width,linestyle='-',color='skyblue',label='sales')
    plt.bar(df['month'],df['profit'],width=width,linestyle='--',color='green',label='profit')
    plt.grid(True)
    plt.legend()


  elif plot_type == 'pie chart':
    fig.set_size_inches(7,5)
    labels = df['month']  # Make sure this exists in your DataFrame
    plt.pie(df['profit'], labels=labels, autopct='%1.1f%%')
    plt.title('Profit Distribution')
    plt.legend(title='Months')

  elif plot_type == 'scatter plot':
    plt.scatter(df['sales'],df['profit'],edgecolors='black',s='100',marker='o')
    plt.xlabel('sales')
    plt.ylabel('profit')
    plt.title('sales vs profit')
    plt.grid(True)

  elif plot_type == 'Histogram':
    plt.hist(df['sales'],bins=5,color='purple',edgecolor='black')
    plt.xlabel('sales')
    plt.ylabel('frequency')
    plt.title('sales distribution')
    plt.grid(True)


  elif plot_type == 'Box plot':
    plt.boxplot(df['sales'],vert=False,patch_artist=True,boxprops=dict(facecolor='lightgreen'))
    plt.xlabel('sales')
    plt.ylabel('profit')
    plt.title('sales distribution')


  plt.tight_layout()
  plt.show()
  return fig


demo=gr.Interface(
    fn=generate_plot,
    inputs=gr.Radio(['line plot','stacked bar chart','pie chart','scatter plot','Histogram','box plot'],
                    label="Choose the plot type"),
    outputs="plot",
    title='sales & profit visual Insight',
    description='choose the type to visualize the data'

)
demo.launch()
demo.launch(share=True)


```
